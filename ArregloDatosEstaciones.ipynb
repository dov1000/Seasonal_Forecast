```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
Data = pd.read_csv('Lluvia_mensual.csv')
```

```python
Data.head()
```

```python
Arenal= Data[['Mes','Year','Arenal']]
Arenal = Arenal.pivot('Year','Mes','Arenal')
```

```python
org=[]
for nombre in Data.columns[2:]:
    x = Data[['Mes','Year',nombre]]
    x = x.pivot('Year','Mes',nombre)
    org.append(x)
Arenal= org[0]
Penas = org[1]
Toro1 = org[2]
Cariblanco = org[3]
RioMacho= org[4]
Cachi = org[5]
Angostura = org[6]
Reventazon = org[7]
Ventanas=org[8]
Pirris = org[9]
```

```python
Penas.to_csv('Penas.csv', header = False)
Arenal.to_csv('Arenal.csv', header = False)
Toro1.to_csv('Toro1.csv', header = False)
Cariblanco.to_csv('Cariblanco.csv', header = False)
RioMacho.to_csv('RioMacho.csv', header = False)
Cachi.to_csv('Cachi.csv', header = False)
Angostura.to_csv('Angostura.csv', header = False)
Reventazon.to_csv('Reventazon.csv', header = False)
Ventanas.to_csv('Ventanas.csv', header = False)
Pirris.to_csv('Pirris.csv', header = False)
```
